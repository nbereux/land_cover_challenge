```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from lcc import OUTPUT_DIR 
```

```python
df = pd.read_csv(OUTPUT_DIR.joinpath('metrics_unet_weighted_ce.csv'))
x = df[df['key'] == 'train_loss']['step']
y_train = df[df['key'] == 'train_loss']['value']
y_test = df[df['key'] == 'test_loss']['value']

kl_div_ce_weighted = df[df['key'] == 'test_kl_div']['value']
jaccard_ce_weighted = df[df['key'] == 'test_jac_idx']['value']

plt.plot(x, y_train, label='Train loss')
plt.plot(x, y_test, label='Test loss')
plt.title('Loss for UNet with weighted cross entropy loss')
plt.legend()
plt.savefig(OUTPUT_DIR.joinpath('loss_unet_weighted_ce.png'))
```

```python
df = pd.read_csv(OUTPUT_DIR.joinpath('metrics_unet_jaccard_loss.csv'))
x = df[df['key'] == 'train_loss']['step']
y_train = df[df['key'] == 'train_loss']['value']
y_test = df[df['key'] == 'test_loss']['value']

kl_div_jaccard = df[df['key'] == 'test_kl_div']['value']
jaccard_jaccard = df[df['key'] == 'test_jac_idx']['value']


plt.plot(x, y_train, label='Train loss')
plt.plot(x, y_test, label='Test loss')
plt.title('Loss for UNet with Jaccard loss')
plt.legend()
plt.savefig(OUTPUT_DIR.joinpath('loss_unet_jaccard_loss.png'))
```

```python
plt.plot(range(len(kl_div_ce_weighted)), kl_div_ce_weighted, label='Weighted CE loss')
plt.plot(range(len(kl_div_jaccard)), kl_div_jaccard, label='Jaccard loss')
plt.legend()
plt.title('KL Divergence for UNet')
plt.savefig(OUTPUT_DIR.joinpath('kl_div_unet.png'))
```

```python
plt.plot(range(len(jaccard_ce_weighted)), jaccard_ce_weighted, label='Weighted CE loss')
plt.plot(range(len(jaccard_jaccard)), jaccard_jaccard, label='Jaccard loss')
plt.legend()
plt.title('Mean Jaccard Index for UNet')
plt.savefig(OUTPUT_DIR.joinpath('jaccard_unet.png'))
```
